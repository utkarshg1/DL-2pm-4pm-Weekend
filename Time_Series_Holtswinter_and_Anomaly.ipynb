### Read dataset

```python
import pandas as pd
df = pd.read_csv("/content/infy_stock.csv", index_col="Date", parse_dates=True)
df.head()
```

```python
df_close = df[['Close']]
df_close
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
Q = seasonal_decompose(df_close['Close'], period=30)
```

```python
Q.observed.plot()
Q.trend.plot()
```

```python
Q.seasonal.plot()
```

```python
Q.resid.plot()
```

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

fit1 = Holt(df_close['Close'],exponential=True, initialization_method="estimated").fit(
    smoothing_level=0.8, smoothing_trend=0.2, optimized=False
)
```

```python
fcast1 = fit1.forecast(60)
fcast1
```

```python
fcast1.plot()
```

### Anomaly detection

```python
df
```

```python
df = df.drop(columns=['Symbol','Series'])
df
```

```python
df.dtypes
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(random_state=0)
clf.fit(df)
```

```python
preds = clf.predict(df)
```

```python
preds
```

```python
df['Anomaly'] = preds
```

```python
df
```

```python
df[df['Anomaly']==-1]
```

```python
df_close
```

```python
clf2 = IsolationForest(random_state=0)
clf2.fit(df_close)
```

```python
preds2 = clf2.predict(df_close)
```

```python
preds2
```

```python
df_close['Anomaly'] = preds2
```

```python
df_close
```

```python
df_close['Close'].plot()
```

```python
import seaborn as sns
sns.lineplot(x=df_close.index, y=df_close['Close'], hue=df_close['Anomaly'],
             palette='rainbow')
```

1. -1 is anomaly
2. 1 is normal data
