### Step 1: Read the dataset

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/main/Mall_Customers.csv')
df.head()
```

### Problem statement : Catgorize similar customers based on Age, Income and spending (Customer Segmentation)

### Step 2 - Perform Basic Data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3 : Seperate X values (Age, Income, Spending)

```python
X1 = df[['Age', 'Income', 'Spending']]
X1.head()
```

### Step 4: Preprocess the data with pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X1)
X_pre.head()
```

### Step 5: Find the optimal number of clusters using yellowbrick package

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

```python
visualizer = KElbowVisualizer(KMeans(),
                              k=(2, 20),
                              timings=False)
visualizer.fit(X_pre)
visualizer.show()
```

### Based on above graph elbow found at k=6 , Create K-means with 6 clusters

```python
km_model = KMeans(n_clusters=6)
lb1 = km_model.fit_predict(X_pre)
```

```python
lb1
```

### Save above value in X1

```python
X_lab1 = X1.copy()
X_lab1['Label'] = lb1
```

```python
X_lab1
```

```python
X_lab1.query("Label==0")
```

```python
X_lab1.query("Label==1")
```

```python
for i in range(6):
  print(f'Label : {i}\n')
  display(X_lab1.query(f"Label=={i}"))
  print('\n=======================================\n')
```

### Get the centroids

```python
cent1 = X_lab1.groupby(by='Label').mean(numeric_only=True)
cent1
```

### Plot above results

```python
import plotly.express as px
fig = px.scatter_3d(data_frame=X_lab1,
                    x = 'Age',
                    y = 'Income',
                    z = 'Spending',
                    color='Label')
fig.show()
```

# Agglomerative Clustering

```python
X_pre
```

```python
from sklearn.cluster import AgglomerativeClustering
ag_model = AgglomerativeClustering(n_clusters=6)
lab2 = ag_model.fit_predict(X_pre)
lab2
```

```python
X_lab2 = X1.copy()
X_lab2['Label'] = lab2
```

```python
X_lab2
```

```python
print('Agglomerative Clustering')
for i in range(6):
  print(f'Label : {i}\n')
  display(X_lab2.query(f"Label=={i}"))
  print('\n==================================================\n')
```

```python
cent2 = X_lab2.groupby(by='Label').mean(numeric_only=True)
cent2
```

```python
fig2 = px.scatter_3d(data_frame=X_lab2,
                     x='Age',
                     y='Income',
                     z='Spending',
                     color='Label')
fig2.show()
```
